## cargar la imagen y convertirla a escala de grises

```python

imagen = cv2.imread('zoee.png')
gris = cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)
```

## encontrar los rostros en la imagen

```python

dr = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
rectRostros = dr.detectMultiScale(gris, scaleFactor=1.2,minNeighbors=5, minSize=(30,30))
print("Se encontr {} rostro(s)".format(len(rectRostros)))

```

```python
for(x , y, w, h) in rectRostros:
    cv2.rectangle(imagen,(x,y),(x+w,y+h),(0,255,0),2)
    
imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
plt.imshow(imagen)
plt.title("Zoe")
plt.show
```
